Levando os dados da aula 01 para essa aula


```python
import pandas as pd

url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'

dados = pd.read_csv(url_dados, compression = 'zip')
dados
```

Vamos efetuar os ajustes na base sugeridos na aula dois

```python
## Retirando o hifem dos nomes das series
names = dados.columns.tolist()
names = [item.replace("-", "") for item in names]
dados.columns = names
dados.head(20)
```

```python
## Alterando no nome da serie "Droga" par "Composto"
mapa = {'droga':'composto'}
dados.rename(columns=mapa, inplace=True)
```

```python
dados.head()
```

```python
cod_compostos =  dados['composto'].value_counts().index[0:5]
dados.query('composto in @cod_compostos' )
cod_compostos
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
plt.figure(figsize=(10, 8))
sns.countplot(x = 'composto', data=dados.query('composto in @cod_compostos' ))
```
